```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graphs

```python

plt.figure(figsize=(4, 3)) # 4:3 aspect ratio

x=[1,2,3]
y=[2,4,6]

plt.title('Simple plot' , fontdict={'fontname':'Comic Sans MS', 'fontsize':15})
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot(x,y)
plt.show()
```

```python
plt.figure(figsize=(4, 3)) # 4:3 aspect ratio
plt.plot(x,y, label='2x', color='red', linewidth=2, marker='o', markersize=10, markeredgecolor='blue', linestyle='--')
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(4, 3)) # 4:3 aspect ratio

# Shorthand notatio fmt = '[color][marker][line]'

plt.plot(x,y, 'r.--', label='2x') # red, dashed line, circle marker

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(4, 3)) 

# Line 2
# Select interval we want to plot points at
x2 = np.arange(0, 4.5, 0.5)
plt.plot(x,y, 'r.--', label='2x') 
plt.plot(x2, x2**2, 'b', label='x^2')
plt.show()
```

```python
plt.figure(figsize=(4, 3))

plt.plot(x,y, 'r.--', label='2x')

# Plot part of the graph as line
plt.plot(x2[:7], x2[:7]**2, 'b', label='x^2')

# Plot remainder of graph as a dot
plt.plot(x2[5:], x2[5:]**2, 'b--')

plt.legend()
plt.show()

#plt.savefig('mygraph.png')
```

### Barchart

```python
plt.figure(figsize=(4, 3))

labels = ['A', 'B', 'C']
values = [1, 4, 2]

bars = plt.bar(labels, values)
plt.show()
```

```python
plt.figure(figsize=(4, 3))

bars = plt.bar(labels, values)
bars[1].set_hatch('/')
plt.show()
```

```python
plt.figure(figsize=(4, 3))

bars=plt.bar(labels, values)
patterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(patterns.pop())
```

```python
plt.figure(figsize=(4, 3))

bars=plt.bar(labels, values)
patterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
```

## Real World Examples

### Line Graph and Gas Price Dataset

```python
plt.figure(figsize=(7, 3))

gas=pd.read_csv('gas_prices.csv')

plt.plot(gas.Year, gas.France, 'b.-', label='France')
plt.plot(gas['Year'], gas['UK'], 'r.-', label='UK')

plt.title('Gas Prices over Time (in USD)', fontdict={'fontname':'Comic Sans MS', 'fontsize':15})
plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.xticks(gas.Year[::3].tolist()+[2011])

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6, 3))
plt.xticks(gas.Year[::3].tolist()+[2011])

# Another way to plot many values!
for country in gas:
    if country != 'Year':
        plt.plot(gas.Year, gas[country], marker='.',label=country)
        
# Adding legend outside the plot
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Adjusting layout to fit the legend
plt.tight_layout()
```

```python
plt.figure(figsize=(5, 3))
plt.xticks(gas.Year[::3].tolist())

# Another way to plot many values!
countries_to_look_at = ['Australia', 'USA', 'France', 'UK']
for country in gas:
    if country in countries_to_look_at:
        plt.plot(gas.Year, gas[country], marker='.',label=country)
        plt.legend()

plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()
```

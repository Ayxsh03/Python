```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('ComputerSales.csv')

# Seaborn provides built in datasets
print(sns.get_dataset_names())

# Load a built in dataset based on US State car crash percentages
crash_df = sns.load_dataset('car_crashes')
```

```python
sns.displot(crash_df['not_distracted'],kde=True)
```

```python
sns.jointplot(x='speeding',y='alcohol',data=crash_df,kind='reg')
```

```python
sns.jointplot(x='speeding',y='alcohol',data=crash_df,kind='kde')
```

### KDE Plots

```python
sns.kdeplot(crash_df['alcohol'])
```

### Pairplot

```python
sns.pairplot(crash_df)
```

```python
tips = sns.load_dataset('tips')

sns.pairplot(tips,hue='sex',palette='coolwarm')
```

### Rug Plot

```python
sns.rugplot(tips['total_bill'])
```
